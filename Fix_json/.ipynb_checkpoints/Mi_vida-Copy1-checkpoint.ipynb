```python
#image = fields[0]
def open_img(name):
    os.chdir('Task1_Train')
    try:
        f = open(name,'r')
        os.chdir('..')
        return f
    except:
        os.chdir('..')
        return []
```

```python
def open_data(name):
    os.chdir('Task2_Train')
    try:
        with open(name) as json_file:
            data = json.load(json_file)
            company = (data['company'])
            date = (data['date'])
            address = (data['address'])
            total = (data['total'])
        os.chdir('..')
        return data
    except:
        os.chdir('..')
        return []
```

```python
def get_new_line(file):
    os.chdir('Task1_Train')
    try:
        l = file.readline()
        length = len(l)
        l = l[:length-1]
        os.chdir('..')
        return l
    except:
        os.chdir('..')
        return []
```

```python
def get_number(l):
    m = re.search('^(.+?),', l)
    n1 = m.group(0)
    length = len(n1)
    l = l[length:]
    n1 = n1[:length-1]
    n1
    return n1, l
```

```python
def get_vectors(li):
    n = []
    for element in range(8):
        n_i, li = get_number(li)
        n.append(n_i)
    return li, [n[2], n[0], n[5], n[1]]
```

```python
def get_id_list(list_i, field_id):
    n_ids = len(list_i)
    t_array = np.array(list(range(n_ids)))+field_id
    field_id = n_ids + field_id
    return t_array.tolist(), field_id
```

```python
#data['company'] == lt
#data['company'] in lt
#text_string = data['company']
def backwards_search(matching_list, numbers_array, text_string):
    #print('the matching list is: ')
    #print(matching_list)
    #print('the numbers array is: ')
    #print(numbers_array)
    for element in range(len(matching_list)):
        lt = ''
        for phrase in matching_list[len(matching_list)-element-1:]:
            lt = lt + ' ' + phrase
        lt = lt[1:]
        #print()
        #print('...............................')
        #print('building_backwards: ' + lt)
        #print()
        matching = lt
        if text_string == matching:
            #print('return by backwards search')
            #print(lt)
            #print(element)
            #print(data['company'] == str(lt))
            return True, numbers_array[len(matching_list)-element-1:], matching_list[len(matching_list)-element-1:]
        for letter in matching:
            matching = matching[1:]
            if text_string == matching:
                return True, numbers_array[len(matching_list)-element-1:], matching_list[len(matching_list)-element-1:]
    return False, [], []
```

```python
def forward_search(file, text_to_match):
    match_completed = False
    l=[]
    n=[]
    while not match_completed:
        
        #try:
            li, ni = get_vectors(get_new_line(file))
            l.append(li)
            n.append(ni)
            if li in text_to_match:
                #print(li + ' was found in ' + text_to_match)
                if li == text_to_match:
                    return True, [ni], [li]
                else:
                    #print(l)
                    status, nr, lr = backwards_search(l, n, text_to_match)
                    if status:
                        return status, nr, lr
            else:
                for letter in li:
                    li = li[1:]
                    if li == text_to_match:
                        return True, [ni], [li]
        #except:
            #print('some shite happened, aborting...')
            #return False, []
```

```python
def prepare(document_name):
    field_id = 1
    text_id = 1
    fields_i=['company', 'date', 'address', 'total']
    json_template = {}
    json_template['global_attributes']={'file_id': document_name[:12] + '.jpg'}
    json_template['fields']=[]
    json_template['text_boxes']=[]
    
    try:

        for inner_field in fields_i:
            flag1, n1, l1 = forward_search(open_img(document_name), open_data(document_name)[inner_field])
            id_list, field_id = get_id_list(l1, field_id)

            json_template['fields'].append(
             {
                'field_name':inner_field,
                 'key_id':[],
                 'key_text':[],
                 'value_id': id_list,
                 'value_text': l1,
             }
               )
            for i_d, bbox, text in zip(id_list, n1, l1):
                json_template['text_boxes'].append(
                 {
                'id': i_d,
                'bbox': bbox,
                'text': text,
                 })

        with open('data/'+document_name[:12] + '.json','w') as outfile:
            json.dump(json_template, outfile)
            copyfile('Task1_Train/'+document_name[:12]+'.jpg', 'data/'+document_name[:12] + '.jpg')
            return True
    except:
            return False
```
